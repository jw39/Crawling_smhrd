#### 경로 설정
1. 절대 경로 : 외부 서버 (제공하고 있는 경로)나 내부에서도 사용 가능
2. 상대 경로 : 현재 위치를 기반으로 찾아가는 경로, 내가 만든 (내부 서버) 파일을 찾아갈 때 쓸 수 있는 경로

- ./        : 현재 위치를 기준으로 (생략 가능)
- / (맨앞에) : 최상단 위치를 기준으로  
- / (중간에) : 폴더 안으로 들어가겠다 
- ../        : 폴더 밖으로 나가겠다

```python
import requests as req
from bs4 import BeautifulSoup as bs
```

# 이번 예제에서 알아가야 할 내용!
1) 응답이 200으로 올바르게 완료가 됐는데, 수집이 안되는 경우 해결책 -> iframe 확인, 절대경로 찾아내기
2) 선택자를 한 번에 가지고 오는 방법 -> 주의점!, 단수개 해결하기

```python
# 1. 리퀘스트를 통해 조회하고자 하는 사이트 링크 보내기
res = req.get("https://finance.naver.com/marketindex/exchangeList.naver")
```

```python
# 2. beatiful soup을 통해 res 안에 있는 text(문자)를 lxml (html)형태로 바꿔주기
soup = bs(res.text, "lxml")
```

```python
# 3. 선택자 가져오기
# f12 -> 요소 -> 마우스 우클릭 -> copy -> copy selector

# 주의점!!! 선택자를 반드시 하나만 가지고 온다, 미국usd만 가져오는 겨... 
# 복수개의 선택자를 수정할 때는 id, nth-child() 지워주기! -> 반드시 선택자는 뒤에서부터 수정한다.
# nth-child() -> li, tr 이 놈들만 건들면 된다

# 긍께 단수 삭제하고 td.tit의 a를 가지고 와야 전체 목록이 가져와짐 
# soup.select("body > div > table > tbody > tr:nth-child(1) > td.tit > a")
# 단수를 의미하는 :nth-child(1) 이 놈을 그냥 삭제
soup.select("body > div > table > tbody > tr> td.tit > a")
```

# iframe 데이터 조회방법
- 선택자, res[200]으로 아무런 문제가 없는 경우 반드시 iframe을 의심해라
- iframe = 다른 서버에 저장된 데이터를 불러다가 보여만 주는 역할
- src 안에 경로가 상대 경로인 경우는 반드시 절대 경로로 변경이 필요함!!!, iframe 안 링크 더블 클릭 후 구글에 쳐보면 알 수 있음
- 상대 경로와 그 페이지인 절대 경로를 보고 iframe의 절대 경로를 파악할 수 있다!..

### 크롤링할 때 사용하는 선택자는 4가지 뿐!!!!
1) 태그 선택자 -> 태그 이름을 작성하는 것 ( h1, a ) => 복수
2) 클래스 선택자 -> 클래스 값을 작성하는 것 (.abc, .title) => 복수
3) 아이디 선택자 -> 아이디값을 작성하는 것 (#id) -> 단수
4) 계층 선택자(자식) -> ">" 를 통해서 활용하는 것
   - nth-child(n) = 몇 번째 자식 -> 단수개 의미
